```python
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from utils.config import settings

# Step 1: Load the SentenceTransformer model
model = SentenceTransformer(settings.MODEL_NAME)

# Step 2: Load the documents from the 'data/raw' folder
def load_documents(path):
    docs = []
    for filename in os.listdir(path):
        if filename.endswith(".txt"):
            with open(os.path.join(path, filename), 'r') as f:
                docs.append(f.read())
    return docs

docs = load_documents(settings.DATA_DIR)

# Step 3: Create embeddings for each document
document_texts = docs
embeddings = model.encode(document_texts, show_progress_bar=True)

# Step 4: Normalize the embeddings (for cosine similarity)
embeddings = embeddings / np.linalg.norm(embeddings, axis=1, keepdims=True)

# Step 5: Build FAISS index
dimension = embeddings.shape[1]  # Dimension of embeddings
index = faiss.IndexFlatIP(dimension)  # Using inner product for cosine similarity
index.add(embeddings)  # Add the embeddings to the index

# Step 6: Save the FAISS index and document IDs
faiss.write_index(index, settings.INDEX_PATH)
np.save("embeddings/doc_ids.npy", np.array([i for i in range(len(docs))]))

print("Index and embeddings saved successfully!")
```

```python
# Step 1: Load the FAISS index and document IDs
index = faiss.read_index(settings.INDEX_PATH)
doc_ids = np.load("embeddings/doc_ids.npy")

# Step 2: Define the query
query = "AI and machine learning advancements"

# Step 3: Encode the query
query_embedding = model.encode([query])

# Step 4: Search the FAISS index for the top 3 closest documents
k = 3
distances, indices = index.search(query_embedding, k)

# Step 5: Display the results
for i, (doc_id, distance) in enumerate(zip(indices[0], distances[0])):
    print(f"Result {i+1}: Document {doc_id} with distance {distance}")

    # Adjust document filename to match actual files in the 'data/raw' folder
    document_filename = os.path.join(settings.DATA_DIR, f"document_{doc_id+1}.txt")

    # Check if the document file exists before trying to open it
    if os.path.exists(document_filename):
        with open(document_filename, 'r') as f:
            print(f"Content: {f.read()}\n")
    else:
        print(f"Error: {document_filename} not found.\n")
```
